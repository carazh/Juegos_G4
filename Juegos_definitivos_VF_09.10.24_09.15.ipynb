```python
import random

def Bienvenido_juegos():

    #solicitar nombre de jugador e iniciar el juego
    nombre_jugador = input("por favor ingresa tu nombre, para empezar a jugar!")
      
    print("**************************")
    print(f"Hola {nombre_jugador} Bienvenido@ a la ronda de juegos de PPT, PR O A")
    print("preparate para una una ronda de experiencia divertidas")
    print("que comience el juego...!Buena Suerte¡")
    print("**************************")

#llamar a la funcion de bienvenida
Bienvenido_juegos()

def elegir_juego():
    
    while True:
            print("Elige uno de los siguientes juegos")
            print("Ahorcado")
            print("Preguntas y respuestas")
            print("Piedra, papel o tijera")
            print("salir")
            elección = input("Introduce el nombre del juego").strip().lower()
            print("**************************")
            if elección == "ahorcado":            
                juegoa_ahorcado()
                print("**************************")
            elif elección == "preguntas y respuestas":
                juego_preguntas_respuestas()
                print("**************************")
            elif elección == "piedra, papel o tijera":
                juego_piedra_papel_tijera()
                print("**************************")
            elif elección == "salir":
                print("Gracias por jugar. ¡Nos vemos pronto!")
                print("**************************")
            else:
                print("Opción no válida")
                print("**************************")

#eleccion del juego. se crea una funcion para ostrar los juegos
#se muestra el menú de juegos disponibles
#ejecución del juego
#elegir_juego()
            seguir_jugando = input("¿Quieres seguir jugando? si/no").strip().lower()
            if seguir_jugando != "si":
                print("¡Gracias por jugar, esperamos volver a verte pronto!")
                break

def juegoa_ahorcado():

    palabra_secreta = "gato"
    lista_secreta = list(palabra_secreta)
    #print("Esta es la lista de elementos de la palabra secreta", lista_secreta)
    intentos_restantes = 6
    #crear lista tablero
    tablero = '*'*len(lista_secreta)
    lista_tablero = list(tablero)

    def dibujar_ahorcado(intentos):
        ahorcado = [
         '''
           +---+
               |
               |
               |
              ===''',
        '''
           +---+
           |   |
               |
               |
              ===''',
        '''
           +---+
           |   |
           O   |
               |
              ===''',
        '''
           +---+
           |   |
           O   |
           |   |
              ===''',
        '''
           +---+
           |   |
           O   |
          /|   |
              ===''',
        '''
           +---+
           |   |
           O   |
          /|\\  |
              ===''',
        '''
           +---+
           |   |
           O   |
          /|\\  |
          /    ===''',
        '''
           +---+
           |   |
           O   |
          /|\\  |
          / \\  ==='''
    ]
    
        print(ahorcado[intentos])


#mientras no se agoten los intentos. 
    while intentos_restantes > 0:
         #para que el usuario ponga una letra. el ptoblema aqui era que en vez de letra adivinada que es la letra del imput. Poniamos adivinar que no es nada
        print("¡Adivina la palabra!:", " ".join(lista_tablero))
        letra_adivinada = input("adivina la palabra").lower()

    #almacenamos posicion de la letra introducida en caso de que la letra esté en la palabra secreta
        if letra_adivinada.lower() in lista_secreta:
            print(f"¡correcto! la letra '{letra_adivinada}' esta en la palabra.")

            #se actualiza el tablero en todas las posiciones donde aparece la letra
            for i in range(len(lista_secreta)):
                if lista_secreta[i] == letra_adivinada:
                    lista_tablero[i] = letra_adivinada
                                   
        else: # en esta parte se reducen los intentos y se dibuja el ahorcado
            intentos_restantes -= 1
            print("oh fallaste")    
            print(f"te quedan '{intentos_restantes}'intentos")
            dibujar_ahorcado(6 - intentos_restantes)
        


        #DUDA! No esta haciendo esta parte. ¿por que? 
        if lista_secreta == lista_tablero:
           print(f"Enhorabuena has ganado, la palabra secreta era {palabra_secreta} !!!")
           break 



def juego_preguntas_respuestas():
#Se crea diccionario con clave: preguntas, valor: respuestas.
    preguntas_respuestas= {"¿Cuál es la ciudad más poblada del mundo?": "Tokio", 
                       "¿Cuál es el desierto más grande del mundo?": "Sahara",
                       "¿Cuál es la montaña más alta de España?": "Teide",
                       "¿En qué país hay un mayor número de pirámides?": "Sudan",
                       "¿Cuál es la cordillera de montañas más larga del mundo?": "Los Andes",
                       "¿Cuál es la isla más grande del mundo?": "Groenlandia",
                       "¿Cuál es el país más grande del mundo por superficie?": "Rusia",
                       "¿Cuál es la capital de Brasil?": "Brasilia",
                       "Cuál es el río más largo del mundo?": "Nilo",
                       "¿Cuál es la montaña más alta del planeta?": "Everest"}

#Se crean variables con contadores y límites de aciertos y fallos.
    contador_aciertos = 0
    contador_fallos = 0
    limite_aciertos = 5
    limite_fallos = 3
#Se crea un bucle while True para que el código se ejecute hasta que se cumplan las condiciones para que el juego se termine.
    while True:
    #Se crea variable con lista de keys del diccionario (las preguntas) que se generarán de manera aleatoria.
        preguntas_clave = list(preguntas_respuestas.keys())
    #Se crea bucle while para la variable pregunta_clave que genera preguntas aleatorias usando un random.choice.
        while preguntas_clave:
            pregunta_aleatoria = random.choice(preguntas_clave)
        #Se hace un print de la pregunta aleatoria para verla en pantalla.
            print(pregunta_aleatoria)
        #Se borra la pregunta aleatoria que se acaba de generar de preguntas_clave con un remove(), de ahí que antes se pasara a lista
        #para que la pregunta no se repita justo despues. 
        #Se siguen generando preguntas aleatorias con el bucle.
            preguntas_clave.remove(pregunta_aleatoria)
        #Se le pide a la usuaria que ingrese su respuesta.
            respuesta_usuaria = input("Ingresa tu respuesta: ")
        #Se introduce una condición para comprobar los aciertos y los fallos. 
        #El strip() y el lower() son código defensivo para que no genere un fallo en el caso de que la usuaria introduzca caracteres que no 
        #coincidan con la respuesta.
        #Si la respuesta de la usuaria se corresponde con el valor del diccionario para la pregunta aleatoria generada, el contador de aciertos suma 1.
            if respuesta_usuaria.strip().lower() == preguntas_respuestas[pregunta_aleatoria].lower():
                contador_aciertos += 1
                print(f"¡Tu respuesta es correcta! Tienes {contador_aciertos} acierto/s.")
        #Si no se corresponde, se suma 1 al contador de fallos.
            else:
                contador_fallos += 1
                print(f"¡Esa respuesta no es correcta, sigue intentándolo! Tienes {contador_fallos} fallos/s.")
        #Se introduce una condición para informar a la usuaria de que ha ganado o perdido según las oportunidades que tiene para jugar.
            if contador_aciertos == limite_aciertos:
                print(f"¡FELICIDADES! ¡HAS GANADO!")
                break
            elif contador_fallos == limite_fallos:
                print(f"¡Oh no! Has perdido. ¡Inténtalo de nuevo!")
                break
    #Se introduce una condición para que el bucle deje de generar preguntas si se llega al límite de aciertos o se llega al límite de fallos. 
        if contador_aciertos == limite_aciertos or contador_fallos == limite_fallos:
            break 

def juego_piedra_papel_tijera():
  # Opciones a elegir.
  opciones = ['piedra', 'papel', 'tijera']
  # Contador de puntos
  ptos_jugador = 0
  ptos_computadora = 0
  puntos_objetivo = 3
  # Es el bucle para llevar registro de ptos y continuar el juego hasta que se alcancen los ptos objetivos = 3
  # tambien se ha determinado el ganador de cada ronda
  while ptos_jugador < puntos_objetivo and ptos_computadora < puntos_objetivo:
      jugador = input("elige piedra, papel o tijera:").lower()
      print(f"El jugador eligió {jugador}")
      computadora = random.choice(opciones)
      print(f"La computadora eligió {computadora}")
      resultado = 0
      # Comparador de elecciones para determinar si el jugador gana la ronda
      if jugador == computadora:
        print("es un empate!")
        print(f"Contador de puntos -- > Jugador: {ptos_jugador} --- > Computadora: {ptos_computadora}")
      elif (jugador == 'piedra' and computadora == 'tijera') or \
        (jugador == 'papel' and computadora == 'piedra') or \
        (jugador == 'tijera' and computadora == 'papel'):
          ptos_jugador +=1
          print("¡Ganaste esta ronda!")
          print(f"Contador de puntos -- > Jugador: {ptos_jugador} --- > Computadora: {ptos_computadora}")
      else:
          ptos_computadora +=1
          print("¡Perdiste! La computadora gana esta ronda")
          print(f"Contador de puntos -- > Jugador: {ptos_jugador} --- > Computadora: {ptos_computadora}")
          # break. Este break no debería ir para que siga el bucle.
      # Determinar quien gana el juego. El primero que llegue a 3 puntos.
      if ptos_jugador == puntos_objetivo:
        print("HAS GANADO")
        break
      elif ptos_computadora == puntos_objetivo:
        print("HA GANADO LA COMPUTADORA")
        break
elegir_juego()
```
